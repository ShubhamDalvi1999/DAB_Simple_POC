# NYC Taxi Analytics

This notebook performs analytics on NYC Taxi trip data.


```python
# Get parameters from job
dbutils.widgets.text("catalog", "dab_first_project")
dbutils.widgets.text("schema", "default")

catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")

print(f"Using catalog: {catalog}, schema: {schema}")
```

```python
# Read NYC Taxi trips data
trips_df = spark.read.table("samples.nyctaxi.trips")
print(f"Total trips: {trips_df.count():,}")
trips_df.display()
```

```python
# Basic analytics - Average fare by pickup zone
from pyspark.sql.functions import avg, sum, count

zone_stats = trips_df.groupBy("pickup_zip").agg(
    count("*").alias("trip_count"),
    avg("fare_amount").alias("avg_fare"),
    sum("fare_amount").alias("total_fare")
).orderBy("total_fare", ascending=False)

zone_stats.display()
```
